```python
"""
1. Write a python program to group the user ids corresponding to the following age group.
• less or equal to 20.
• between 20 and 40.
• more than 40.
"""

lessthan20 = []
between20and40 = []
morethan40 = []

for i in data:
    if i['Age'] < 20:
        lessthan20.append(i['user id'])
    elif i['Age'] > 40:
        morethan40.append(i['user id'])
    else:
        between20and40.append(i['user id'])

print("The user ids that have age less than 20 are",lessthan20)
print("The user ids that have age between 20 and 40 are",between20and40)
print("The user ids that have age more than 40 are",morethan40)
```

```python
"""
2.Create a line chart for user ids and weights. User id on the x-axis and weight on the y-axis.
"""

import matplotlib.pyplot as plt

userids = []
weights = []

for i in data:
    userids.append(i['user id'])
    weights.append(i['Weight'])

plt.plot(userids, weights)
plt.xlabel("User id")
plt.ylabel("Weight")
plt.title("User id VS Weight")
plt.show()
```

```python
"""
3. Write a python program to create a list of 20 vectors by taking user’s height, weight and age. Find the
sum of all these 20 vectors. After that, find the average height, average weight and average age.
Hint: [150,60,23] will be one vector.
"""

vectors = []

for i in data:
    vector = []
    vector.append(i['Height'])
    vector.append(i['Weight'])
    vector.append(i['Age'])
    vectors.append(vector)

print("The Vectors are",vectors)

totalWeight = 0
totalHeight = 0
totalAge = 0
count = 0
for i in vectors:
    totalHeight += i[0]
    totalWeight += i[1]
    totalAge += i[2]
    count += 1

print('The average weight is',totalWeight/count)
print('The average height is',totalHeight/count)
print('The average age is', totalAge/count)
```

```python
"""
4. Write a python program to find the mean, median and mode of weights of users.
"""

weights = []

for i in data:
    weights.append(i['Weight'])

totalWeight = sum(weights)
totalpeople = len(weights)

print("The mean weight is ", totalWeight/totalpeople)

weights.sort()

if totalpeople % 2 == 0:
    median1 = weights[totalpeople//2]
    median2 = weights[totalpeople//2 - 1]
    median = (median1 + median2)/2
else:
    median = weights[totalpeople//2]

print("The median weight is ", median)

frequencyWeight = {}

for i in weights:
    if i in frequencyWeight:
        frequencyWeight[i] = frequencyWeight.get(i)+1
    else:
        frequencyWeight[i] = 1

modes = []
max=0

for k,v in frequencyWeight.items():
    if v == max:
        modes.append(k)
    elif v > max:
        max = v
        modes.clear()
        modes.append(k)

print("The mode of the weight is:",end=' ')
for i in modes:
    print(i , end=' ')
```

```python
"""
5. Write a python program to find the variance of heights of users.
"""
import statistics


meanHeight = totalHeight/totalpeople

deviation = []

for i in data:
    deviation.append(i['Height']-meanHeight)

deviation = [i**2 for i in deviation]
sumSquareDeviation = sum(deviation)
variance = sumSquareDeviation/(totalpeople-1)

print("The variance of heights of users is" , variance)

#Alterate method to calculate variance
heights=[]
for i in data:
    heights.append(i['Height'])
variance = statistics.variance(heights)
print("The alternative variance of heights of users is" , variance)
```

```python
"""
6. Write a python program to find covariance between weights and heights of users.
"""

import numpy as np

weights=[]
heights=[]

for i in data:
    weights.append(i['Weight'])
    heights.append(i['Height'])

covarience = np.cov(weights, heights)[0,1]
print("The covariance between weights and heights of users is ", covarience)


#Alternative Method

meanWeight = totalWeight/totalpeople
meanHeight = totalHeight/totalpeople

weights = [i-meanWeight for i in weights]
heights = [i-meanHeight for i in heights]


sum = 0
for i in range(len(weights)):
    sum += weights[i]*heights[i]

covarience = sum/(totalpeople-1)

print("The alternative way for covariance between weights and heights of users is ", covarience)


```
