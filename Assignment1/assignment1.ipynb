```python
#Ashutosh Rath - 2041018066
#Question 1 and 2

salaries_and_tenure = [(83000,8.7),(88000,8.1),(48000,0.7),(76000,6),(69000,6.5),(76000,7.5),(60000,2.5),(83000,10),(48000,1.9),(63000,4.2)]

g1 = [] #Less then two
g2 = [] #Between two and five
g3 = [] #More than five

for s,t in salaries_and_tenure:
    if t<2 :
        print(f'({s},{t}) --> Less than two')
        g1.append(s)
    elif t >=2 and t<=5 :
        print(f'({s},{t}) --> More than two and Less than five')
        g2.append(s)
    else:
        print(f'({s},{t}) --> More than five')
        g3.append(s)

print("Average salary for less than two years of tenure: ", sum(g1) / len(g1))
print("Average salary for between two years and five years of tenure: ", sum(g2) / len(g2))
print("Average salary for more than five years of tenure: ", sum(g3) / len(g3))
```

```python
import matplotlib.pyplot as plt

sal = [s[0] for s in  salaries_and_tenure]
ten = [s[1] for s in salaries_and_tenure]

plt.title("Salary vs Year of Experience")
plt.xlabel("Experience(yrs)")
plt.ylabel("Salary($)")
plt.scatter(ten,sal)
plt.grid()

```

```python
import random
import string

def generate(length):
    str = string.digits +  string.ascii_letters
    password = ''.join(random.choices(str,k=length))
    return password

length = int(input("Enter the length of the password to be generated"))
no_of_password = int(input("Enter the number of the password to be generated"))

password = [generate(length) for i in range(no_of_password)]
print(password)
```
